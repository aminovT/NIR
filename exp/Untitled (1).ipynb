```python
import matplotlib.pyplot as plt
from numpy import load
import numpy as np 
```

```python
prier = ['experiments_with_DP_95_' , "experiments_with_DP_50_" , 'experiments_without_bias' , 'experiments']
mod = ['lin_connect' , 'arc_connect']
p_f =  ['DP_95' , "DP_50" ,  'without', 'with_bias' ]
pf = ['PointFinderWithBias', 'PointFinderInverseWithBias' , 'PointFinderTransportation']
pr = ['' , 'WA' , 'OT']
hd = [  50, 100 , 500, 1000, 2000]
lin = np.linspace(0,1,21)
for j in mod:
    for p in pf:
        for k in hd:
            plt.figure(figsize=(15,10))
            plt.title(("model=" + pr[pf.index(p)] + j + str(k)))
            plt.grid()
            for i in prier:
                path = '{}/eval/LinearOneLayer{}/{}{}/curve.npz'.format(i,k ,p,j)
                data = load(path)
                plt.plot(lin, data['tr_err'] , label = (p_f[prier.index(i)]))
            plt.legend()
            plt.show()
                
                
```
