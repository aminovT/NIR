```python
import matplotlib.pyplot as plt
```

```python
mod = ['lin_connect' , 'arc_connect']
pf = ['PointFinderWithBias'] #, 'PointFinderInverseWithBias' , 'PointFinderTransportation']
hd = [ 10, 50, 100 , 500, 1000, 2000]

for m in mod:
    plt.figure()
    plt.title(m)
    plt.grid()
    for p in pf:
        for i in hd:
            path = 'experiments/eval/LinearOneLayer{}/{}{}/curve.npz'.format(i,p,m)
            data = load(path)
            plt.plot(data['tr_err'] , label = str(i))
            
    plt.legend()
    plt.show()
```
