```python
import matplotlib.pyplot as plt
from numpy import load
import numpy as np 
```

```python
prier = ['experiments_with_DP_95_' , "experiments_with_DP_50_" , 'experiments_without_bias' , 'experiments']
mod = ['lin_connect' , 'arc_connect']
p_f =  ['DP_95' , "DP_50" ,  'clear', 'normal' ]
pf = ['PointFinderWithBias', 'PointFinderInverseWithBias' , 'PointFinderTransportation']
pr = ['' , 'WA+' , 'OT+']
md = ['lin' , 'arc']
hd = [  50, 100 , 500, 1000, 2000]
lin = np.linspace(0,1,21)
for j in mod:
    for p in pf:
        for k in hd:
            plt.figure()
            s = "model=" + pr[pf.index(p)] + md[mod.index(j)] + ' (hidden = {})'.format(k)
            plt.title((s))
            plt.grid()
            for i in prier:
                path = '{}/eval/LinearOneLayer{}/{}{}/curve.npz'.format(i,k ,p,j)
                data = load(path)
                plt.plot(lin, data['tr_err']/min( data['tr_err']) , label = (p_f[prier.index(i)]))
            plt.xlabel('ВРЕМЯ')
            plt.ylabel('ОШИБКА')         
            plt.legend(loc = 1 , prop={'size': 12})
            plt.savefig(fname= 'imgg/'+s)
            plt.show()
                
                
```
